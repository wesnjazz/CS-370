# Numpy.roll

```python
x = np.arange(10)
np.roll(x, 3)
```

```python
x2 = np.reshape(x, (2, 5))
x2
```

```python
np.roll(x2, 1)
```

```python
np.roll(x2, 1, axis=1)
```

```python
np.roll(x2, 1, axis=0)
```

```python
c = np.random.randint(0, 255, (5, 5))
c
```

```python
c[::2, ::2]
```

```python
d= np.random.randint(0, 255, (5, 5))
d

```

```python
d = np.zeros((5, 5))
d
```

```python
d[::2, ::2] = c[::2, ::2]
d
```

```python
d[1::2, 1::2] = c[1::2, 1::2]
d
```

```python
d[1::2, 0::2] = c[1::2, 0::2]
d
```

```python
d[0::2, 1::2] = c[0::2, 1::2]
d
```

```python
e = np.random.randint(0, 255, (5, 5))
e
```

```python
f = np.zeros((5, 5))
f
```

```python
f[::2, ::2] = e[::2, ::2]
f
```

```python
f[1::2, 1::2] = e[1::2, 1::2]
f
```

```python
f[1::2, 0::2] = e[1::2, 0::2]
f
```

```python
f[0::2, 1::2] = e[0::2, 1::2]
f
```
