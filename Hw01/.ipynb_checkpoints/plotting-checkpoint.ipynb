```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('data/demosaic/puppy.jpg')
```

```python
# RED
plt.figure(figsize=(10,8))
plt.scatter(img[::2, 1::2, 0], img[::2, ::2, 0], c='r')
plt.title('Red Plotting')
plt.savefig('PlottingRed.png')
plt.show()
```

```python
# GREEN
plt.figure(figsize=(10,8))
plt.scatter(img[::2, 1::2, 1], img[::2, ::2, 1], c='g')
plt.title('Green Plotting')
plt.savefig('PlottingGreen.png')
plt.show()
```

```python
# BLUE
plt.figure(figsize=(10,8))
plt.scatter(img[::2, 1::2, 2], img[::2, ::2, 2], c='b')
plt.title('Blue Plotting')
plt.savefig('PlottingBlue.png')
plt.show()
```
